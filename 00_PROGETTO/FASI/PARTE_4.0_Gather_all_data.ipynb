```python
import pandas as pd

ristoranti_analizzati = 20

path_1 = '/content/gdrive/My Drive/Colab Notebooks/CLEAN_DATA_FOR_ANALYSIS/RISTORANTE_'
path_2 = '.xlsx'

# Salvo il nome dei ristoranti in una lista
all_files = []
for i in range(ristoranti_analizzati):
  path = path_1 + str(i) + path_2
  all_files.append(path)

# Li rendo compatibili con pandas
data = []
for file in all_files:
    data.append(pd.read_excel(file))

# concateno i dati
ALL_DATA = pd.concat(data, ignore_index=True)
```

```python
display(ALL_DATA)
```

## Salvo il file

```python
!pip install xlsxwriter
import pandas as pd

# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('TUTTI_I_RISTORANTI.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
ALL_DATA.to_excel(writer, sheet_name='Sheet1')

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```

```python
ALL_DATA_SORTED = ALL_DATA.sort_values(by=['Voti'])

display(ALL_DATA_SORTED[:10])
```

```python
ALL_DATA_SORTED['Index'] = range(1, len(ALL_DATA_SORTED) + 1)

display(ALL_DATA_SORTED)
```

```python
ALL_DATA_TRAINING_SET = ALL_DATA_SORTED.loc[(ALL_DATA_SORTED.Index % 2 == 0)]

display(ALL_DATA_TRAINING_SET)

# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('ALL_DATA_TRAINING_SET.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
ALL_DATA_TRAINING_SET.to_excel(writer, sheet_name='Sheet1')

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
